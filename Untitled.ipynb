```python
import rioxarray as rxr
```

```python
data = rxr.open_rasterio('USGS_13_n20w156.tif', masked=True)
data
```

```python
data[0].rio.to_raster('test.tif')
```

```python
data.plot()
```

```python
geo = [
    {
        'type': 'Polygon',
        'coordinates': [[
            [-155.6,19.8],
            [-155.4,19.8],
            [-155.4,19.6],
            [-155.6,19.6],
            [-155.6,19.8]
        ]]
    }
]
```

```python
clipped = data.rio.clip(geo)
```

```python
clipped
```

```python
clipped.plot()
```

```python
clipped[0].plot.contour(levels=20)
```

```python
clipped2 = clipped.rio.clip_box(minx=-155.5,miny=19.775,maxx=-155.45,maxy=19.8)
```

```python
clipped2.plot()
```

```python
clipped2.rio.to_raster('sample.tif')
```
